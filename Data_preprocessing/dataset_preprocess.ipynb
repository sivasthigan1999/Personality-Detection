# Two pair combination

```python
import numpy as np

# Sample correlation matrix
matrix = [
    [1, 0.332, 0.129, 0.049, 0.270],
    [0.332, 1, 0.240, 0.031, 0.507],
    [0.129, 0.240, 1, 0.253, 0.403],
    [0.049, 0.031, 0.253, 1, 0.074],
    [0.270, 0.507, 0.403, 0.074, 1]
]

traits = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']

correlation_pairs = []

# Extract combinations and avoid redundancy
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        correlation_pairs.append((traits[i], traits[j], matrix[i][j]))

# Sort based on absolute correlation values
sorted_pairs = sorted(correlation_pairs, key=lambda x: abs(x[2]), reverse=True)

# Get top 10
top_10 = sorted_pairs[:10]

for pair in top_10:
    print(pair)
```

# Three pair combination

```python
import numpy as np

# Sample correlation matrix
matrix = [
    [1, 0.332, 0.129, 0.049, 0.270],
    [0.332, 1, 0.240, 0.031, 0.507],
    [0.129, 0.240, 1, 0.253, 0.403],
    [0.049, 0.031, 0.253, 1, 0.074],
    [0.270, 0.507, 0.403, 0.074, 1]
]

traits = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']

correlation_triplets = []

# Extract combinations for three traits
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        for k in range(j+1, len(matrix)):
            correlation_triplets.append((traits[i], traits[j], traits[k], matrix[i][j], matrix[i][k], matrix[j][k]))

# Sort based on average absolute correlation values
sorted_triplets = sorted(correlation_triplets, key=lambda x: (abs(x[3]) + abs(x[4]) + abs(x[5])) / 3, reverse=True)

# Get top 10
top_10 = sorted_triplets[:10]

for triplet in top_10:
    print(triplet)
```

# 2+3 combination

```python
import numpy as np

# Sample correlation matrix
matrix = [
    [1, 0.332, 0.129, 0.049, 0.270],
    [0.332, 1, 0.240, 0.031, 0.507],
    [0.129, 0.240, 1, 0.253, 0.403],
    [0.049, 0.031, 0.253, 1, 0.074],
    [0.270, 0.507, 0.403, 0.074, 1]
]

traits = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']

correlations = []

# Extract two-pair combinations
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        average_corr = abs(matrix[i][j])
        correlations.append((traits[i], traits[j], average_corr))

# Extract three-pair combinations
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        for k in range(j+1, len(matrix)):
            average_corr = (abs(matrix[i][j]) + abs(matrix[i][k]) + abs(matrix[j][k])) / 3
            correlations.append((traits[i], traits[j], traits[k], average_corr))

# Sort based on average absolute correlation values
sorted_correlations = sorted(correlations, key=lambda x: x[-1], reverse=True)

# Get top 10
top_10 = sorted_correlations[:10]

for entry in top_10:
    print(entry)
```

```python
import numpy as np

# Sample correlation matrix
matrix = [
    [1, 0.332, 0.129, 0.049, 0.270],
    [0.332, 1, 0.240, 0.031, 0.507],
    [0.129, 0.240, 1, 0.253, 0.403],
    [0.049, 0.031, 0.253, 1, 0.074],
    [0.270, 0.507, 0.403, 0.074, 1]
]

traits = ['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']

correlations = []

# Extract two-pair combinations
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        average_corr = abs(matrix[i][j])
        correlations.append((traits[i], traits[j], average_corr))

# Extract three-pair combinations
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        for k in range(j+1, len(matrix)):
            average_corr = (abs(matrix[i][j]) + abs(matrix[i][k]) + abs(matrix[j][k])) / 3
            correlations.append((traits[i], traits[j], traits[k], average_corr))

# Extract four-pair combinations
for i in range(len(matrix)):
    for j in range(i+1, len(matrix)):
        for k in range(j+1, len(matrix)):
            for l in range(k+1, len(matrix)):
                average_corr = (abs(matrix[i][j]) + abs(matrix[i][k]) + abs(matrix[i][l]) + abs(matrix[j][k]) + abs(matrix[j][l]) + abs(matrix[k][l])) / 6
                correlations.append((traits[i], traits[j], traits[k], traits[l], average_corr))

# Sort based on average absolute correlation values
sorted_correlations = sorted(correlations, key=lambda x: x[-1], reverse=True)

# Get top 10
top_10 = sorted_correlations[:10]

for entry in top_10:
    print(entry)
```

```python
import pandas as pd
from langdetect import detect, DetectorFactory

# Seed to make langdetect deterministic
DetectorFactory.seed = 0
def is_english(sentence):
    try:
        return detect(sentence) == 'en'
    except:
        return False

# Define a function to count words in a string
def word_count(string):
    return len(string.split())

# Define a threshold for minimum number of words
MIN_WORD_COUNT = 15
  # Change this based on your requirement

# Filter rows based on the word count in the 'STATUS' column
df = df[df['STATUS'].apply(word_count) >= MIN_WORD_COUNT]
df_filtered = df[~df['STATUS'].str.contains('[/*;^&<>@]', regex=True)]
df_english = df_filtered[df_filtered['STATUS'].apply(is_english)]




```

```python
import pandas as pd
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
# Load the model and tokenizer
model_name = "gpt2-medium"  # You can use "gpt2" for the smaller model or other variants for larger ones
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def sentence_probability(sentence):
    input_ids = tokenizer.encode(sentence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
    log_likelihood = outputs[0].item()
    return log_likelihood

# Load your dataset
df = pd.read_csv('/content/filtered_dataset_15withsymbolremoval01.csv')

# Compute sentence likelihood and filter out unlikely sentences
# Note: the threshold is subjective and should be adjusted based on your data and needs
threshold = -50  # This is just a sample value; adjust as needed based on your observations
filtered_df = df[df['STATUS'].apply(sentence_probability) > threshold]

# If you want to save the filtered dataframe
# filtered_df.to_csv('filtered_dataset.csv', index=False)
```
